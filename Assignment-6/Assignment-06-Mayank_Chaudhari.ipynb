### Problem-1

A startup company with 20 employees recorded their annual salaries (in INR Lakhs):
`[6.5, 7.2, 8.0, 6.8, 9.5, 7.0, 15.0, 7.5, 8.2, 6.9, 7.8, 10.5, 7.1, 8.5, 6.7, 7.3, 7.9, 9.0, 8.8, 7.6]`

(a) Calculate the mean, median, and mode of the salaries.

(b) Determine the range, variance, and standard deviation of the salaries.

(c) Calculate the Interquartile Range (IQR). What does the IQR tell you about the salary distribution?

(d) Construct a box plot for the salaries. Identify any potential outliers.


```python
# Imports
import numpy as np
import statistics
import matplotlib.pyplot as plt

# Data (Annual Salaries in INR Lakhs)
salaries = [
    6.5, 7.2, 8.0, 6.8, 9.5, 7.0, 15.0, 7.5, 8.2, 6.9,
    7.8, 10.5, 7.1, 8.5, 6.7, 7.3, 7.9, 9.0, 8.8, 7.6
]

# Convert to NumPy array for computation
arr = np.array(salaries)
```

```python
# Mean
mean_salary = np.mean(arr)

# Median
median_salary = np.median(arr)

# Mode (handle case where no unique mode exists)
try:
    mode_salary = statistics.mode(salaries)
except statistics.StatisticsError:
    mode_salary = None

print("a) Central Tendency")
print("Mean:", round(mean_salary, 2))
print("Median:", round(median_salary, 2))
print("Mode:", mode_salary if mode_salary else "No unique mode")
```

```python
# Range
salary_range = np.max(arr) - np.min(arr)

# Sample Variance (ddof=1)
variance_salary = np.var(arr, ddof=1)

# Sample Standard Deviation
std_dev_salary = np.std(arr, ddof=1)

print("\n---- (b) Dispersion ----")
print("Range:", round(salary_range, 2))
print("Sample Variance:", round(variance_salary, 2))
print("Sample Standard Deviation:", round(std_dev_salary, 2))
```

```python
# First and Third Quartiles
Q1 = np.percentile(arr, 25)
Q3 = np.percentile(arr, 75)

# IQR
IQR = Q3 - Q1

# Outlier boundaries (1.5 Ã— IQR rule)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = arr[(arr < lower_bound) | (arr > upper_bound)]

print("\n---- (c) IQR Analysis ----")
print("Q1:", round(Q1, 2))
print("Q3:", round(Q3, 2))
print("IQR:", round(IQR, 2))
print("Lower Bound:", round(lower_bound, 2))
print("Upper Bound:", round(upper_bound, 2))
print("Outliers:", outliers if len(outliers) > 0 else "None")
```

```python
plt.boxplot(arr)
plt.title("Salary Distribution (INR Lakhs)")
plt.ylabel("Salary (INR Lakhs)")
plt.show()
```

### Problem-2

A tech company conducted a customer satisfaction survey, rating their latest product on a scale of 1 to 10 (10 being highly satisfied). The scores from 30 randomly selected customers are:
[7, 8, 9, 6, 7, 8, 10, 5, 7, 8, 9, 7, 6, 8, 7, 9, 10, 6, 7, 8, 5, 9, 8, 7, 6, 7, 8, 9, 7, 8]

(a) Create a frequency distribution table for these scores.

(b) Plot a histogram of the customer satisfaction scores. Describe the shape of the distribution (e.g., symmetric, skewed).

(c) Calculate the Pearson's (first and second) coefficients of skewness for this data. Interpret the values.

(d) Calculate the excess kurtosis for this data using sample kurtosis formula. What does the kurtosis value suggest about the tails of the distribution?

```python

from collections import Counter

# Customer satisfaction scores
scores = [
    7, 8, 9, 6, 7, 8, 10, 5, 7, 8,
    9, 7, 6, 8, 7, 9, 10, 6, 7, 8,
    5, 9, 8, 7, 6, 7, 8, 9, 7, 8
]

arr = np.array(scores)
n = len(arr)
```

```python
# Frequency distribution
frequency = Counter(scores)

print("Score | Frequency")
print("------------------")
for score in sorted(frequency):
    print(f"{score:^5} | {frequency[score]:^9}")
```

```python
plt.hist(arr, bins=range(5, 12))
plt.title("Customer Satisfaction Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.xticks(range(5, 11))
plt.show()

mean_val = np.mean(arr)
median_val = np.median(arr)

if mean_val > median_val:
    print("Distribution is positively skewed (right-skewed)")
elif mean_val < median_val:
    print("Distribution is negatively skewed (left-skewed)")
else:
    print("Distribution is approximately symmetric")
```

```python
mean_val = np.mean(arr)
median_val = np.median(arr)
std_val = np.std(arr, ddof=1)

# Mode
mode_val = max(frequency, key=frequency.get)

# Pearson's first coefficient
pearson_first = (mean_val - mode_val) / std_val

# Pearson's second coefficient
pearson_second = 3 * (mean_val - median_val) / std_val

print("\n--- Pearson's Skewness ---")
print("Mean:", round(mean_val, 3))
print("Median:", round(median_val, 3))
print("Mode:", mode_val)
print("Std Dev:", round(std_val, 3))
print("Pearson First Coefficient:", round(pearson_first, 3))
print("Pearson Second Coefficient:", round(pearson_second, 3))
```

```python
mean_val = np.mean(arr)
std_val = np.std(arr, ddof=1)

# Standardized values
z = (arr - mean_val) / std_val

# Apply sample excess kurtosis formula
term1 = (n * (n + 1)) / ((n - 1) * (n - 2) * (n - 3))
term2 = np.sum(z**4)
term3 = (3 * (n - 1)**2) / ((n - 2) * (n - 3))

excess_kurtosis = term1 * term2 - term3

print("\n--- Excess Kurtosis ---")
print("Excess Kurtosis:", round(excess_kurtosis, 3))
```

|    10 |                     8 |                72 |


(a) Calculate the covariance between advertising spend and sales revenue.

(b) Plot Sales Revenue vs Advertising spent.

(c) Calculate the Pearson correlation coefficient between advertising spend and sales revenue.

(d) Interpret the meaning of both the covariance and the correlation coefficient in the context of this data.

```python


# Advertising Spend (X) and Sales Revenue (Y)
X = np.array([5, 7, 6, 8, 5, 9, 7, 10, 6, 8])
Y = np.array([52, 65, 60, 70, 55, 75, 68, 80, 58, 72])

n = len(X)
```

```python
mean_X = np.mean(X)
mean_Y = np.mean(Y)

covariance = np.sum((X - mean_X) * (Y - mean_Y)) / (n - 1)

print("Sample Covariance:", round(covariance, 3))
```

```python
std_X = np.std(X, ddof=1)
std_Y = np.std(Y, ddof=1)

correlation = covariance / (std_X * std_Y)

print("Pearson Correlation Coefficient:", round(correlation, 4))
```

```python
if correlation > 0:
    print("There is a positive linear relationship between advertising spend and sales revenue.")
elif correlation < 0:
    print("There is a negative linear relationship between advertising spend and sales revenue.")
else:
    print("There is no linear relationship between advertising spend and sales revenue.")

print("\nInterpretation:")
print("- Positive covariance indicates that as advertising increases, sales tend to increase.")
print("- Correlation close to +1 indicates a strong positive linear relationship.")
```

### Problem-4

A traffic department claims that the average daily commute time for residents in a certain city is 45 minutes. A researcher believes this claim is too low and collects data from a random sample of 100 residents. The sample mean commute time is 48 minutes with a standard deviation of 12 minutes.

(a) Formulate the null and alternative hypotheses for this scenario.

(b) Construct a 95% confidence interval for the true average daily commute time.

(c) Based on the confidence interval, what can you conclude about the traffic department's claim?

Since the researcher believes the claim is too low, this is a right-tailed test.

H<sub>0</sub>â€‹: $\mu$=45

H<sub>1</sub>â€‹: $\mu$>45

Considering it a larger sample
sample size = 100
We can use Z distribution

For 95% confidence: Z0.025â€‹=1.96

```python

from scipy import stats

# Given values
mu_0 = 45    
sample_mean = 48
sample_std = 12
n = 100
confidence_level = 0.95

# (a) Hypotheses

print("Null Hypothesis (H0): Î¼ = 45")
print("Alternative Hypothesis (H1): Î¼ > 45")


# (b) 95% Confidence Interval

alpha = 1 - confidence_level
z_critical = stats.norm.ppf(1 - alpha/2)

standard_error = sample_std / np.sqrt(n)
margin_of_error = z_critical * standard_error

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print("\n for 95% Confidence Interval ")
print("Lower Bound:", round(lower_bound, 2))
print("Upper Bound:", round(upper_bound, 2))

# (c) Conclusion


if mu_0 < lower_bound:
    print("\nConclusion: 45 minutes is below the confidence interval.")
    print("Reject H0. The true mean commute time is likely greater than 45 minutes.")
elif mu_0 > upper_bound:
    print("\nConclusion: 45 minutes is above the confidence interval.")
    print("Reject H0.")
else:
    print("\nConclusion: 45 minutes lies within the confidence interval.")
    print("Fail to reject H0.")
```

Since 45 is below the lower bound (45.65): we reject the null hypothesis

```python

# Machine Production Analysis
import math
# Given values
sample_mean = 150
sample_std = 15
n = 36
confidence_level = 0.99


# (a) Point Estimate


print("Point Estimate of Mean:", sample_mean)


# (b) 99% Confidence Interval (t-distribution)


alpha = 1 - confidence_level
df = n - 1

# t critical value
t_critical = stats.t.ppf(1 - alpha/2, df)

standard_error = sample_std / np.sqrt(n)
margin_of_error = t_critical * standard_error

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print("\n--- 99% Confidence Interval ---")
print("Lower Bound:", round(lower_bound, 2))
print("Upper Bound:", round(upper_bound, 2))


# (c) Required Sample Size for Margin = 3


desired_margin = 3

# For planning, use Z critical value
z_critical = stats.norm.ppf(1 - alpha/2)

required_n = (z_critical * sample_std / desired_margin) ** 2
required_n = math.ceil(required_n)  # round up

print("\nRequired Sample Size for Â±3 units margin at 99% confidence:")
print(required_n)
```

### Problem-6

A website designer believes that a new website redesign will increase the average time users spend on the site. Before the redesign, the average time spent was 3.5 minutes. After the redesign, a random sample of 50 users showed an average time of 3.8 minutes with a standard deviation of 1.0 minute.

(a) Perform a one-sided hypothesis test at a 0.05 significance level to determine if the redesign significantly increased the average time spent.

(b) State your null and alternative hypotheses, calculate the test statistic, and determine the p-value.

(c) What is your conclusion regarding the website redesign's impact?

Claimed mean -> $\mu_0$ = 3.5

Sample mean -> $\bar{x}$  = 3.8

Sample std deviation -> s = 1.0

Sample size -> n = 50

Significance level -> $\alpha$ = 0.05

a) time increased, this is a right-tailed test. 

H<sub>0</sub> â€‹: $\mu$ =3.5

H1â€‹:$\mu$ >3.5

b) Test Statistic & p-value

$t = \frac{\bar{x} - \mu_0}{\frac{s}{\sqrt{n}}}$

Standard Error:  SE=1/50  â‰ˆ 0.1414

Test Statistic: t=(3.8âˆ’3.5)/0.1414â‰ˆ2.12

p-value â‰ˆ 0.019

c) Since : p<0.05

We reject Hâ‚€.







```python

# Website Redesign Analysis

# Given values
mu_0 = 3.5      
sample_mean = 3.8
sample_std = 1.0
n = 50
alpha = 0.05


# (a) Hypotheses

print("Null Hypothesis (H0): Î¼ = 3.5")
print("Alternative Hypothesis (H1): Î¼ > 3.5")

# (b) Test Statistic

standard_error = sample_std / np.sqrt(n)

t_stat = (sample_mean - mu_0) / standard_error

df = n - 1

# One-sided p-value
p_value = 1 - stats.t.cdf(t_stat, df)

print("\nTest Statistic (t):", round(t_stat, 4))
print("Degrees of Freedom:", df)
print("P-value:", round(p_value, 6))

# Decision Rule

if p_value < alpha:
    print("\nDecision: Reject the null hypothesis.")
else:
    print("\nDecision: Fail to reject the null hypothesis.")
```

### Problem-8

A battery manufacturer claims that their new 'ProLife' battery lasts longer than their existing 'Standard' battery. A consumer watchdog organization tests 40 'ProLife' batteries and 50 'Standard' batteries. The results are:

ProLife: Sample mean life = 120 hours, Sample standard deviation = 10 hours

Standard: Sample mean life = 115 hours, Sample standard deviation = 12 hours

(a) State the null and alternative hypotheses for testing the manufacturer's claim.

(b) Perform a t-test for two independent means at a 0.02 significance level.

(c) What is your conclusion regarding the manufacturer's claim about the 'ProLife' battery?

a) hypothesis :

$\mu_1$ = Mean life of ProLife battery

$\mu_2$ = Mean life of Standard battery

manufacturer claims ProLife lasts longer ; 

H0â€‹:$\mu_1$â€‹ â‰¤ $\mu_2$

H1â€‹:$\mu_1$ â€‹> $\mu_2$

Null Hypothesis: ProLife does not last longer than Standard.

Alternative Hypothesis: ProLife lasts longer than Standard.

This is a right-tailed test.
â€‹

```python
#b) t-test

# Battery Life 

# Given data
mean1 = 120   # ProLife
std1 = 10
n1 = 40

mean2 = 115   # Standard
std2 = 12
n2 = 50

alpha = 0.02


#Compute Test Statistic


se = np.sqrt((std1**2 / n1) + (std2**2 / n2))

t_stat = (mean1 - mean2) / se


# Step 2: Degrees of Freedom 

numerator = (std1**2 / n1 + std2**2 / n2) ** 2
denominator = ((std1**2 / n1)**2 / (n1 - 1)) + ((std2**2 / n2)**2 / (n2 - 1))

df = numerator / denominator


# Step 3: One-Sided p-value

p_value = 1 - stats.t.cdf(t_stat, df)


print("t-statistic:", round(t_stat, 4))
print("Degrees of freedom:", round(df, 2))
print("One-sided p-value:", round(p_value, 6))

if p_value < alpha:
    print("\nDecision: Reject H0")
else:
    print("\nDecision: Fail to Reject H0")

```

Reject the null hypothesis

### Bonus Problem

An e-commerce company is testing two different designs for its product pages (Design A and Design B) to see if one leads to significantly higher conversion rates (percentage of visitors who make a purchase). They randomly assigned 50,000 website visitors to see Design A and another 50,000 visitors to see Design B over a two-week period. At the end of the period, they collected data on the conversion rates. Due to data collection specifics, they have daily average conversion rates for 30 randomly selected days for each design, rather than individual visitor data.

Here are the summary statistics for the daily average conversion rates (in percentage points):

* Design A (Sample 1):

    - Number of days ($n_A$) = 30
    - Mean Daily Conversion Rate ($\bar{x}_A$) = 2.85%
    - Sample Standard Deviation ($s_A$) = 0.45%

* Design B (Sample 2):

    - Number of days ($n_B$) = 30
    - Mean Daily Conversion Rate ($\bar{x}_B$) = 3.05%
    - Sample Standard Deviation ($s_B$) = 0.60%

The company is concerned that the variability in conversion rates might differ between the two designs due to different aesthetic elements or backend processes.

(a) State the null and alternative hypotheses to test if there is a significant difference in the average daily conversion rates between Design A and Design B. (Consider a two-tailed test, as they are looking for any significant difference).

(b) Before performing the t-test, you need to decide whether to assume equal or unequal population variances. For the purpose of this problem, assume that a prior test suggests equal population variances.

(c) Calculate the appropriate test statistic (t-value) for comparing the two independent means, assuming equal population variances. Show all steps of your calculation.

(d) Determine the degrees of freedom for this test. 

(e) Using a significance level of Î±=0.05, determine the critical value(s) for your test.

(f) Based on your calculated test statistic and the critical value(s), what is your conclusion? Is there a statistically significant difference in the average daily conversion rates between Design A and Design B?

(g) Construct a 95% confidence interval for the difference between the true mean daily conversion rates $(Î¼_B - Î¼_A)$. Interpret this interval in the context of the problem.

(h) Show all the steps when a prior test suggests unequal population variances.
