#### 7. What are the top 5 most populated Counties according to the 2010 Census?

```python
# Write your code here
sort_by_2010Census = (pop.groupby(['County','State'], as_index=False)['2010Census'] #Assumptiom  state + county  => one unique county
                      .sum()
                      .nlargest(5,"2010Census"))

print(sort_by_2010Census['County'].to_list()) 
```

#### 8. What are the top 5 most populated States according to the 2010 Census?

```python
# Write your code here

sort_by_2010Census = (pop.groupby('State', as_index=False)['2010Census']
                      .sum()
                      .sort_values("2010Census",ascending=False)
                     )

print(sort_by_2010Census.head(5))
```

#### 9. How many Counties have 2010 populations greater than 1 million?

```python
# Write your code here
county_wise_population = (pop.groupby('County', as_index=False)['2010Census'].sum())

print(county_wise_population[county_wise_population['2010Census']>1000000])
```

#### 10. How many Counties don't have the word "County" in their name?

```python
# Write your code here

counties_without_county = pop[~pop['County'].str.contains('County',case=False)][['County','State']]
print(counties_without_county.to_string(index=False))
```
