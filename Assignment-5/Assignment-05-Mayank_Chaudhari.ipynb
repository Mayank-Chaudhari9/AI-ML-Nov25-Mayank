### Problem-1

In a network, data packets are sent through a shared channel. Suppose there are 10 devices attempting to send data. In any given time slot, each device has a 20% chance of attempting to send a packet, independently of other devices. A collision occurs if two or more devices attempt to send a packet in the same time slot. What is the probability that in a given time slot, there will be no collisions?

````python
import math
p_send = 0.2
n = 10
p_zero = (1 - p_send) ** n
p_one = math.comb(n, 1) * p_send * (1 - p_send) ** (n - 1)
p_no_collision = p_zero + p_one
print(f"Probability : no collision  = {p_no_collision:.6f}")
````

### Problem-3

A software project has two modules, A and B. The probability that module A has a bug is 0.15, and the probability that module B has a bug is 0.10. The probability that both modules have bugs is 0.03.

(a) What is the probability that module A has a bug, given that module B has a bug?

(b) What is the probability that module B has a bug, given that module A does not have a bug?

````python
p_a, p_b, p_ab = 0.15, 0.10, 0.03
p_a_given_b = p_ab / p_b
p_b_given_not_a = (p_b - p_ab) / (1 - p_a)
print(f"probability that module A has a bug, given that module B has a bug : (a) P(A|B) = {p_a_given_b}")
print(f"probability that module A has a bug, given that module B has a bug : (b) P(B|A') = {p_b_given_not_a:.6f}")
````

### Problem-4

A robot uses a sensor to detect an obstacle. The sensor has a 95% chance of correctly detecting an obstacle when one is present (true positive rate). However, it also has a 2% chance of indicating an obstacle when there isn't one (false positive rate).
In the robot's environment, obstacles are present 10% of the time.

What is the probability that an obstacle is truly present, given that the sensor indicates an obstacle?

````python
p_obstacle = 0.10
p_sensor_given_obstacle = 0.95
p_sensor_given_no_obstacle = 0.02
p_sensor = p_sensor_given_obstacle * p_obstacle + p_sensor_given_no_obstacle * (1 - p_obstacle)
p_obstacle_given_sensor = (p_sensor_given_obstacle * p_obstacle) / p_sensor
print(f"probability  obstacle is truly present, given that the sensor indicates obstacle : P(obstacle | sensor indicates) = {p_obstacle_given_sensor:.6f}")
````

(a) What is the probability that exactly 3 microchips in the sample are defective?
(b) What is the probability that at least 2 microchips in the sample are defective?

````python
from scipy import stats
n, p = 20, 0.02
binom = stats.binom(n, p)
p_exactly_3 = binom.pmf(3)
p_at_least_2 = 1 - binom.pmf(0) - binom.pmf(1)
print(f"Probability exactly 3 chips are defective : (a) P(X=3) = {p_exactly_3:.6f}")
print(f"Probability at least 2 chips are defective : (b) P(X≥2) = {p_at_least_2:.6f}")
````

### Problem-8

The time (in hours) until a critical server component fails follows an exponential distribution with a mean of 500 hours.

(a) What is the probability that the component will last between 400 and 600 hours?

(b) The component has already been running for 300 hours without failure. What is the probability that it will last for at least another 250 hours?

````python
from scipy import stats
mean = 500
expon = stats.expon(scale=mean)
p_400_600 = expon.cdf(600) - expon.cdf(400)
p_another_250 = expon.sf(250)  # P(T > 250), memoryless
print(f"Probability the component lasts between 400 and 600 hours : (a) P(400 < T < 600) = {p_400_600:.6f}")
print(f"Probability it lasts at least another 250 hours given it already ran 300 hours : (b) P(T > 250 more hours | already 300) = {p_another_250:.6f}")
````

### Problem-9

(b) If a reading is considered an "outlier" if its noise is more than 1.5 standard deviations away from the mean, what is the probability that a random reading is an outlier?

````python
from scipy import stats
norm = stats.norm(0, 0.5)
p_between = norm.cdf(0.3) - norm.cdf(-0.2)
p_outlier = 2 * (1 - norm.cdf(1.5 * 0.5))  # |Z|>1.5 means |X|>0.75
# Or: p_outlier = 2 * norm.sf(1.5) for standard normal
p_outlier_std = 2 * stats.norm.sf(1.5)
print(f"Probability noise is between −0.2 and 0.3 : (a) P(-0.2 < noise < 0.3) = {p_between:.6f}")
print(f" Probability a reading is an outlier (> 1.5σ from mean) : (b) P(outlier) = {p_outlier_std:.6f}")
````

### Problem-10

(a) What is the probability that a data packet experiences a latency greater than 180 ms?

(c) If a network engineer defines "critical latency" as any latency exceeding 2.5 standard deviations above the mean, what is the probability that a packet experiences critical latency?

````python
from scipy import stats
mu, sigma = 150, 20
norm = stats.norm(mu, sigma)
p_gt_180 = norm.sf(180)
threshold_90 = norm.ppf(0.90)
p_critical = norm.sf(150 + 2.5 * 20)
print(f" Probability latency is greater than 180 ms : (a) P(latency > 180 ms) = {p_gt_180:.6f}")
print(f" Latency threshold for 90% of packets : (b) 90% threshold = {threshold_90:.2f} ms")
print(f" Probability of critical latency (> 2.5σ above mean) : (c) P(cr itical latency) = {p_critical:.6f}")
````

### Problem-12
The probability density function (pdf) of the triangular distribution is characterized by three points as shown in the figure below.

<img src="https://upload.wikimedia.org/wikipedia/commons/4/45/Triangular_distribution_PMF.png" alt="triangular distribution" width="500"/>

(a) min value of the distribution,
(b) max value of the distribution, and
(c) mode of the distribution

Note: $a \leq c \leq b$

Create a class `TriangularDistribution` with attributes `min`, `max` and `mode`. You need to implement the following methods:

- `pdf(x)`: returns the value of the pdf at point $x$.
- `plot_pdf()`: plots the pdf in the range $[min , max]$
- `cdf(x)`: returns the value of the cdf at point $x$ ; which denotes $P(X \leq x)$
- `plot_cdf()`: plots the cdf in the range $[min , max]$
- `get_probability(x1, x2)`: returns the probability $P(x_1 \leq X \leq x_2)$, where $x_1 \lt x_2$.

Gracefully handle the possible errors and also the edge conditions when either $ a = c $ or $b = c$ in the distribution.

Test your implementation.

````python
### Write your code here
import numpy as np
import matplotlib.pyplot as plt

class TriangularDistribution:
    """Triangular distribution with parameters min (a), max (b), and mode (c). a <= c <= b."""

    def __init__(self, min_val, max_val, mode):
        if not (min_val <= mode <= max_val):
            raise ValueError("Require min <= mode <= max")
        if min_val >= max_val:
            raise ValueError("min must be less than max")
        self.min = float(min_val)
        self.max = float(max_val)
        self.mode = float(mode)
        self._a, self._b, self._c = self.min, self.max, self.mode

    def pdf(self, x):
        """Returns pdf at x. Handles edge cases a=c and b=c."""
        a, b, c = self._a, self._b, self._c
        if x < a or x > b:
            return 0.0
        if a == c:
            if x == a:
                return 2.0 / (b - a) if b > a else 0.0
            return 2.0 * (b - x) / ((b - a) ** 2)
        if b == c:
            if x == b:
                return 2.0 / (b - a) if b > a else 0.0
            return 2.0 * (x - a) / ((b - a) ** 2)
        if x <= c:
            return 2.0 * (x - a) / ((b - a) * (c - a))
        return 2.0 * (b - x) / ((b - a) * (b - c))

    def cdf(self, x):
        """Returns P(X <= x)."""
        a, b, c = self._a, self._b, self._c
        if x < a:
            return 0.0
        if x >= b:
            return 1.0
        if a == c:
            return 1.0 - (b - x) ** 2 / (b - a) ** 2 if x > a else 0.0
        if b == c:
            return (x - a) ** 2 / (b - a) ** 2
        if x <= c:
            return (x - a) ** 2 / ((b - a) * (c - a))
        return 1.0 - (b - x) ** 2 / ((b - a) * (b - c))

    def plot_pdf(self):
        x = np.linspace(self.min, self.max, 500)
        y = [self.pdf(xi) for xi in x]
        plt.figure(figsize=(8, 4))
        plt.plot(x, y, 'b-')
        plt.xlabel('x'); plt.ylabel('pdf(x)')
        plt.title(f'Triangular PDF (min={self.min}, max={self.max}, mode={self.mode})')
        plt.grid(True); plt.show()

    def plot_cdf(self):
        x = np.linspace(self.min, self.max, 500)
        y = [self.cdf(xi) for xi in x]
        plt.figure(figsize=(8, 4))
        plt.plot(x, y, 'g-')
        plt.xlabel('x'); plt.ylabel('cdf(x)')
        plt.title(f'Triangular CDF (min={self.min}, max={self.max}, mode={self.mode})')
        plt.grid(True); plt.show()

    def get_probability(self, x1, x2):
        """Returns P(x1 <= X <= x2). Requires x1 < x2."""
        if x1 >= x2:
            raise ValueError("Require x1 < x2")
        return self.cdf(x2) - self.cdf(x1)
````

````python
tri = TriangularDistribution(0, 2, 1)
print("pdf(0.5) =", tri.pdf(0.5), "| cdf(1) =", tri.cdf(1))
print("P(0.5 <= X <= 1.5) =", tri.get_probability(0.5, 1.5))
tri.plot_pdf()
tri.plot_cdf()
````

```python

```

Write a function `rand7()` that will generate a uniform random integer within the range [1, 7] by only using `rand5()`. 

Note: You can call `rand5()` as many time as you want inside the function `rand7()`, but you can not use `np.random.randint(1, 8)` or any other built-in methods in python. 

````python
### Write your code here
import numpy as np

def rand5():
    return np.random.randint(1, 6)

def rand7():
    """Uniform random integer in [1, 7] using only rand5() (rejection sampling)."""
    while True:
        u = 5 * (rand5() - 1) + (rand5() - 1) #for getting uniform 0...24
        if u < 21:
            return (u % 7) + 1

np.random.seed(42)
samples = [rand7() for _ in range(7000)]
from collections import Counter
counts = Counter(samples)
print("Sample counts for rand7():", dict(sorted(counts.items())))
````

### Problem-14

A grayscale image can be represented as a matrix where the dimensions are (height, width). 
Let's consider a simplified scenario with two grayscale images, $A$ and $B$, both represented as 10×10 matrices where each element represents pixel intensity. Pixel intensity is an integer in the range [0-255].

You want to create a blended image $C$ by combining $A$ and $B$ such that $C=w_1 A+w_2 B$. This represents a weighted average blend.
Note that each element in the resultant matrix $C$ should also fall within the range [0, 255]. 

write the code of the function `blend_images(image_A, image_B, weight_A, weight_B)` that takes two matrices and two scalar weights and produces another matrix as an output by performing the above operation.

````python
### Write your code here
import numpy as np

def blend_images(image_A, image_B, weight_A, weight_B):
    """Blend two grayscale images: C = w1*A + w2*B, clipped to [0, 255]."""
    A = np.asarray(image_A, dtype=float)
    B = np.asarray(image_B, dtype=float)
    if A.shape != B.shape:
        raise ValueError("image_A and image_B must have the same shape")
    C = weight_A * A + weight_B * B
    C = np.clip(C, 0, 255)
    return C.astype(np.uint8)

#Exapmple
np.random.seed(0)
img_A = np.random.randint(0, 256, (10, 10))
img_B = np.random.randint(0, 256, (10, 10))
blended = blend_images(img_A, img_B, 0.6, 0.4)
print("blend_images(10x10, 10x10, 0.6, 0.4) shape:", blended.shape)
print("Output range:", blended.min(), "-", blended.max())
blended
````
